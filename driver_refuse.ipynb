```python
#-*- coding: utf-8 -*-
%matplotlib inline
from matplotlib.font_manager import FontProperties  
import matplotlib.pyplot as plt  
import numpy as np
import pandas as pd
import os
plt.rcParams['font.sans-serif'] = ['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False #用来正常显示负号
```

```python
pieces=[]
columns=['ID','warning','empty_taxi','dome_light'
                        ,'elevated','brake','received_time','GPS_time','Longitude','Latitude','speed','direction','satellite']
for sample in range(4):
    path='/Users/ppj/Desktop/taxi_data/part-0000%d.csv' %sample
    frame=pd.read_csv(path,names=columns)
    pieces.append(frame)
taxi=pd.concat(pieces,ignore_index=True)
```

```python
taxi.GPS_time=taxi.GPS_time.map(lambda x: x[11:])
```

```python
taxi_mor=taxi[(taxi.GPS_time>='07:30:00')&(taxi.GPS_time<='09:30:00')]
taxi_even=taxi[(taxi.GPS_time>='16:30:00')&(taxi.GPS_time<='19:00:00')]
taxi_midnight=taxi[(taxi.GPS_time>='00:00:00')&(taxi.GPS_time<='04:00:00')]
taxi_other=taxi[((taxi.GPS_time>='04:00:00')&(taxi.GPS_time<='07:30:00'))+((taxi.GPS_time>='09:30:00')&(taxi.GPS_time<='12:30:00'))+
               ((taxi.GPS_time>='19:00:00')&(taxi.GPS_time<='23:59:59'))]
```

```python
names=['ID','empty_taxi','dome_light']
taxi_se=taxi.loc[:,names]
taxi_even_se=taxi_even.loc[:,names]
taxi_mor_se=taxi_mor.loc[:,names]
taxi_midnight_se=taxi_midnight.loc[:,names]
taxi_other_se=taxi_other.loc[:,names]
```

```python
ID_unique=taxi_se.ID.unique()
```

```python
def compute_rate(taxi_id):
    taxi_id_filter=taxi_id[taxi_id.empty_taxi==1][taxi_id.dome_light=='0']
    if len(taxi_id)>0:
        return float(len(taxi_id_filter))/len(taxi_id)
    else:
        return np.nan
```

```python
def driver_refuse(taxi_id):
    taxi_id_se=taxi_se[taxi_se.ID==taxi_id]
    taxi_id_even=taxi_even_se[taxi_even_se.ID==taxi_id]
    taxi_id_mor=taxi_mor_se[taxi_mor_se.ID==taxi_id]
    taxi_id_midnight=taxi_midnight_se[taxi_midnight_se.ID==taxi_id]
    taxi_id_other=taxi_other_se[taxi_other_se.ID==taxi_id]
    collect_list=[taxi_id_se,taxi_id_even,taxi_id_mor,taxi_id_midnight,taxi_id_other]
    collect_end=[taxi_id]
    for x in collect_list:
        collect_end.append(compute_rate(x))
    return collect_end
        
```

```python
sample_list=[13468.0,15990.0,20227.0,25439,18204.0,21615,25242.0,25413.0,13411.0,11808]
```

```python
collect_refuse=[]
for ID_time in ID_unique[0:]:
    collect_refuse.append(driver_refuse(ID_time))
```

```python
driver_refuse_info=pd.DataFrame(collect_refuse,columns=['ID','total_refuse','evening_refuse','morning_refuse','midnight_refuse','other_refuse'])
```
